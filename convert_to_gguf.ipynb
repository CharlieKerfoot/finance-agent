```python
from unsloth import FastLanguageModel

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "ckerf/arbagent-llama3-8b-lora",
    max_seq_length = 8192,
    dtype = None,
    load_in_4bit = False,
)

model.save_pretrained_merged("arbagent_merged_16bit", tokenizer, save_method = "merged_16bit")
```
